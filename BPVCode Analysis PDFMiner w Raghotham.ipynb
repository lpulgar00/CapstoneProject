```python
from pdfminer.pdfinterp import PDFResourceManager, PDFPageInterpreter    #process_pdf
from pdfminer.pdfpage import PDFPage
from pdfminer.converter import TextConverter
from pdfminer.layout import LAParams
from six import BytesIO as StringIO

import pandas as pd
import re

```

```python
pdfFile = 'BPVC_I_2017BODYTEST.pdf' # This cell converts the pdf to text file in the rawest of forms.
def pdf_to_text(pdfFile):

    # PDFMiner boilerplate
    rsrcmgr = PDFResourceManager()
    sio = StringIO()
    codec = 'utf-8'
    laparams = LAParams()
    device = TextConverter(rsrcmgr, sio, codec=codec, laparams=laparams)
    interpreter = PDFPageInterpreter(rsrcmgr, device)
    pagenos = set()

    # Extract text
    fp = open(pdfFile, 'rb')
    for page in PDFPage.get_pages(fp, pagenos):  #FIGURE OUT TO EXTRACT THE PAGE NUMBERS
        interpreter.process_page(page)
    fp.close()

    # Get text from StringIO
    text = sio.getvalue()
   
    # Cleanup
    device.close()
    sio.close()
       
    return text

output = pdf_to_text(pdfFile)  # this output is a txt file
```

```python
BPVC_1_txt = open('BPVC_I Conversion_to_text.txt', 'w', errors ='ignore') #this cell write the output of the pdf to txt file into a txt file
#use the split or replace

NewOutput = output.decode('utf-8')

#cleaning txt file attempt 1
NewOutput = NewOutput.replace('\n\n',' ')
NewOutput = NewOutput.replace('-\n', '')
NewOutput = NewOutput.replace('\n', ' ')
# NewOutput
BPVC_1_txt.write(NewOutput)

BPVC_1_txt.close()

```

```python
import spacy
nlp = spacy.load('en_core_web_sm')

sentencizer = nlp.create_pipe('sentencizer')
nlp.add_pipe(sentencizer)

#cleaning for  attempt 2

with open('./BPVC_I Conversion_to_text.txt' ,'r',encoding ='ascii', errors = 'surrogateescape') as f:
    data = f.read()

#cleaning for punctuation attempt 3 look into regex
data = data.replace(',','').replace(':','').replace(';', '').replace(')', '').replace('(', '')  # Removing some punctuation
    
data = data.encode(encoding = 'utf-8', errors = 'ignore')  #encode into bytes


#begin spacy action

spacy_stopwords = spacy.lang.en.stop_words.STOP_WORDS #list/extract the built in stopwords
# print('Num of stop words:',  len(spacy_stopwords))
# print('First ten stop words', list(spacy_stopwords)[:10])

doc = nlp(data.decode())  # decode back into string





```

```python
#Extracting all caps headers
def ExtractKeywords(col):
    return re.findall('[A-Z]{3,}', str(col)) 

df['Keywords'] = df['Sentences'].apply(ExtractKeywords)

```

```python
#extract header e.g. PG-2
def ExtractFirstLevelHeader(col):
    return re.findall('PG-\d+',str(col)) 


df['FirstLevelHeaders'] = df['Sentences'].apply(ExtractFirstLevelHeader)

```

```python
#extract subheader e.g. PG-2.1
def ExtractSubHeader(col):
    return re.findall('PG-\d+.\d+', str(col)) 

df['Subheader'] = df['Sentences'].apply(ExtractSubHeader)

df
```
